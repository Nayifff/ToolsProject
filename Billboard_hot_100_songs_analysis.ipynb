```python
import collections
def get_song_peryear(df): ### displays the number of songs per year being used after cleaning
    df_list = []

    for index, row in df.iterrows():
        df_list.append(row.values.tolist())

    x =[]
    for i in range(len(df_list)):
        x.append(df_list[i][0])

    count_years = collections.Counter(x)
    return (count_years)
get_song_peryear(data4)
```

We are trying to understand the trend of music from 1960s - making use of the "Billboard hot 100" songs as representative. The change is significant. Some notable observations:
1. Danceability - we see a hike in the mean index around the mide 80s(remember Michael Jackson?), which dips down again aound 2000s, acting as a normally distributed curve.
2. Key - We see a dip in the average key over the years. Key of a piece is the group of pitches, or scale, that forms the basis of a music composition in classical, Western art, and Western pop music. We see classical music losing it's charm with the masses nowadays... this is just the proof.
3. Loudness - With Rock bands decreasing day by day - specifically hard metal bands, we see a dip in the mean loudness index of songs.
4. Mode - It depicts whethee the mode was major(1) - signifying upliting mood - or minor(0) - tending more towards the melancholic side. We see that as we move closer to 2018, it keeps on diminshing. 
5. Speechiness - This one has seen it's upside down throughout the years. It sees peak around 2000s, the time when rap music was blooming; affecting the speechiness index!
6. Acousticness, liveliness, instrumentalness - All of them decrease drastically. As time passes by, electronic artists are seizing thrones from teh guitarists.
7. valence - As the index moves towards 1, the song becomes more positive. It is decreasing over time.
8. Tempo - decreases with time.

If we see the pattern, it is around the year 2000 more or less all covariates start to trend in a similar fashion.Hence, it can be set as a threshold for our next goal - preparation of prediction models. We will start our analysis from the year 2000 to analyze further.
